# Part 2

```python
#Query 2.A: Joining Cholesterol and Hypertension to make a single table. 

query_2a = """ 
SELECT 
    h.Region,
    h.Country, 
    h.Year, 
    h.Gender, 
    h.Hypertension_Value, 
    c.Cholesterol_Value 

FROM 
    hypertension_table_1 h
JOIN 
    cholesterol_table_1 c 
ON
    h.Year = c.Year AND h.Gender = c.Gender;
"""

combined_a = pd.read_sql_query(query_2a, engine)
combined_a.head()
      
```

```python
#Query 2.B - Joining the combined Hypertension and Cholesterol data with Mortality data to make a table. 

query_2b = """ 
SELECT 
    h.Region,
    h.Country, 
    h.Year, 
    h.Gender, 
    h.Hypertension_Value, 
    c.Cholesterol_Value,
    m.Mortality_Value,
    m.Disease

FROM 
    hypertension_table_1 h
JOIN 
    cholesterol_table_1 c 
ON
    h.Year = c.Year AND h.Gender = c.Gender
JOIN 
    mortality_table_1 m
ON 
    h.Year = m.Year AND h.Gender = m.Gender;
"""

combined_b = pd.read_sql_query(query_2b, engine)
combined_b.head()

canada_data=pd.read_sql_query(query_2b, engine) 
canada_data.to_sql("canada_data", engine, index=False, if_exists='replace')
canada_data=pd.read_sql_table("canada_data", engine)
canada_data.head()
```

```python
#Query 4.A: Ranking countries based on hypertension (from highest to lowest).  

query_4a = """ SELECT Country, 
    AVG(Hypertension_Value) AS Average_Hypertension_Value,
    RANK() OVER (ORDER BY AVG(Hypertension_Value) DESC) AS 'Rank'
FROM 
    hypertension_table
WHERE 
    Gender = 'Both sexes' 
    AND Year BETWEEN 2000 AND 2018
GROUP BY 
    Country
ORDER BY 
    'Rank';"""              

pd.read_sql_query(query_4a, engine)

query_4a_rank = pd.read_sql_query(query_4a, engine)
query_4a_rank

#Rank is the new column that shows the ranking of the countries based on hypertension value 
```

```python
#Determining the rank of Canada from 2000 to 2018 based on hypertension.

canada_rank = query_4a_rank[query_4a_rank['Country'] == 'Canada']

average_rank = canada_rank['Rank'].mean()
average_value = canada_rank['Average_Hypertension_Value'].mean()

rounded_average_rank = round(average_rank)
rounded_average_value = round(average_value)

print(f"The average rank for Canada based on hypertension from 2000 to 2018 is {rounded_average_rank} with an average value of {rounded_average_value}.")
```

```python
#Query 4.b: Ranking countries based on cholesterol (from highest percentage to lowest percentage).  

query_4b = """ SELECT Country, 
    AVG(Cholesterol_Value) AS Average_Cholesterol_Value,
    RANK() OVER (ORDER BY AVG(Cholesterol_Value) DESC) AS 'Rank'
FROM 
    cholesterol_table
WHERE 
    Gender = 'Both sexes' 
    AND Year BETWEEN 2000 AND 2018
GROUP BY 
    Country
ORDER BY 
    'Rank';"""               

pd.read_sql_query(query_4b, engine)

query_4b_rank = pd.read_sql_query(query_4b, engine)
query_4b_rank
```

```python
canada_rank_b = query_4b_rank[query_4b_rank['Country'] == 'Canada']

average_rank_b = canada_rank_b['Rank'].mean()
average_value_b = canada_rank_b['Average_Cholesterol_Value'].mean()

rounded_average_rank_b = round(average_rank_b)
rounded_average_value_b = round(average_value_b)

print(f"The average rank for Canada based on cholesterol from 2000 to 2018 is {rounded_average_rank_b} with an average value of {rounded_average_value_b}.")
```

```python
#Query 5.A: Taking the average of female and male hypertension values by region. 

query_5a = """ SELECT Region, Gender, 
AVG(Hypertension_Value) AS Average_Hypertension_Value,
RANK() OVER (ORDER BY AVG(Hypertension_Value) ASC) AS 'Rank' 
FROM hypertension_table
WHERE Gender = 'Male' OR Gender = 'Female'
AND Year BETWEEN 2000 AND 2018
GROUP BY Region, Gender
ORDER BY 'Rank'; """

hypertension_avg_data = pd.read_sql_query(query_5a, engine)
hypertension_avg_data.head(15)
```

```python
#Query 5.b: Taking the average of female and male cholesterol values by region.

query_5b = """ SELECT Region, Gender, 
AVG(Cholesterol_Value) AS Average_Cholesterol_Value,
RANK() OVER (ORDER BY AVG(Cholesterol_Value) ASC) AS 'Rank'
FROM cholesterol_table 
WHERE Gender = 'Male' OR Gender = 'Female'
AND Year BETWEEN 2000 AND 2018
GROUP BY Region, Gender 
ORDER BY 'Rank'; """ 

cholesterol_avg_data = pd.read_sql_query(query_5b, engine)
cholesterol_avg_data.head(15)
```

```python
#Query 5.c: Determining the trend of mortality 

query_5c = """ SELECT Region, Gender, 
AVG(Mortality_Value) AS Average_Mortality_Value,
RANK() OVER (ORDER BY AVG(Mortality_Value) ASC) AS 'Rank'
FROM mortality_table 
WHERE Gender = 'Male' OR Gender = 'Female'
AND Disease = 'Cardiovascular diseases'
AND Year BETWEEN 2000 AND 2018
GROUP BY Region, Gender 
ORDER BY 'Rank'; """ 

mortality_trend_data = pd.read_sql_query(query_5c, engine)
mortality_trend_data.head(15)
```
